```python
import numpy as np
from scipy import stats

data = np.random.randint(10, 100, 30)

print("Data:", data)

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", stats.mode(data, keepdims=False))
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

normal_data = np.random.normal(loc=50, scale=10, size=1000)

sns.histplot(normal_data, kde=True, color='skyblue')

plt.title("Normal Distribution of Simulated Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
coin = np.random.choice(['Heads', 'Tails'], size=100)

heads_prob = np.sum(coin == 'Heads') / len(coin)

print("Heads Probability:", heads_prob)
```

```python
dice = np.random.randint(1, 7, size=1000)

for i in range(1, 7):
    prob = np.sum(dice == i) / len(dice)
    print(f"Probability of {i}: {prob:.2f}")
```

```python
from scipy import stats
import numpy as np

group_a = np.random.normal(160, 10, 30)
group_b = np.random.normal(165, 12, 30)

t_stat, p_value = stats.ttest_ind(group_a, group_b)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis: The group means differ significantly.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference.")
```

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

data = np.random.normal(50, 5
                        , 100)
data[5] = 100

z_scores = np.abs(stats.zscore(data))

outliers = np.where(z_scores > 3)

print("Outlier indices:", outliers)

plt.plot(data, marker='o', label='Data Points')
plt.plot(outliers[0], data[outliers], 'ro', label='Outliers')

plt.title("Outlier Detection using Z-score")
plt.legend()
plt.show()
```
